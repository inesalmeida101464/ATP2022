Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = (nlugares, Vendidos, filme)
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
  ...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.


````python
def menu(opcao):
    print()
    print("""1-Filmes em exibição
2- Consultar disponibilidade de lugar
3- Lista lugares ocupados
4- Lugares disponíveis
5-Acrescentar nova sala
0-Terminar programa""")
    return opcao

cinema=[(150,[],"Titanic"),(200,[],"Matrix")]

def listar(cinema):
    listafilmes=[]
    for sala in cinema:
        nlugares,vendidos,filme=sala
        listafilmes.append(filme)
    print()
    print(">>>Filmes em exibição :", listafilmes,"<<<")
    return listafilmes

def disponivel(cinema,lugar,filmec):
    resp=False
    for sala in cinema:
        nlugares,vendidos, filme=sala
        if filme==filmec:
            if lugar not in vendidos and lugar<=nlugares:
                resp=True
            else:
                resp=False
    return resp 

def lugOcupados(cinema,lugar,filmec):
    disponivel(cinema,lugar,filmec)
    acresc=input("Deseja ocupar o lugar?(S/N")
    if (acresc=="S" or acresc== "s"):
        for sala in cinema:
            nlugares,vendidos,filme=sala
            if filmec==filme:
                vendidos.append(lugar)
    elif (acresc=="N" or acresc=="n"):
        return cinema 
    return cinema
    
def lugDisponiveis(cinema):
    listap=[]
    for sala in cinema:
        nlugares,vendidos,filme=sala
        disp=nlugares-len(vendidos)
        listadisp=(filme,disp)
        listap.append(listadisp)
    return listap
    

def novasala(cinema,filmec,lotaçao):
    novasala=(lotaçao,[],filmec)
    cinema.append(novasala)
    return cinema

print("""1-Filmes em exibição
2- Consultar disponibilidade de lugar
3- Lista lugares ocupados
4- Lugares disponíveis
5-Acrescentar nova sala
0-Terminar programa""")
opcao=int(input("Selecione uma opção do menu abaixo"))

while opcao!=0:
    if opcao==1:
        listar(cinema)
        menu(opcao)
        opcao=int(input("Selecione uma opção do menu abaixo"))
    
    if opcao==2:
        filmec=input("Insira o filme")
        lugar=int(input("Insira o lugar"))
        print(disponivel(cinema,lugar,filmec))
        opcao=int(input("Selecione uma opção do menu abaixo"))
        menu(opcao)
        

    if opcao==3:
        print(lugOcupados(cinema,lugar,filmec))
        opcao=int(input("Selecione uma opção do menu abaixo"))
        menu(opcao)
    
    if opcao==4:
        print(lugDisponiveis(cinema))
        opcao=int(input("Selecione uma opção do menu abaixo"))
        menu(opcao)

    if opcao==5:
        filmec=input("Nome de novo filme:")
        lotaçao=int(input("Qual a lotação da sala"))
        print(novasala(cinema,filmec,lotaçao))
        opcao=int(input("Selecione uma opção do menu abaixo"))
        menu(opcao)
    
if opcao==0:
    print("Terminou o programa. De momento as disponibilidades das salas:")
    print(cinema)

        
        
````
